# Filtrando Dataframe do Pandas com Base em Strings

```python
import pandas as pd
import numpy as np
```

```python
dataSet = pd.read_csv('dataset.csv')
```

```python
dataSet.head()
```

```python
# Filtrando coluna por valor inicial

dataSet[dataSet.Segmento.str.startswith('Con')].head()
```

```python
dataSet[dataSet.Categoria.str.endswith('lies')].head()
```

```python
dataSet.Segmento.value_counts()
```

# Split de string em DATAFRAME Pandas

```python
dataSet['ID_Pedido'].head()
```

```python
# Dividindo as informações de 'ID_Pedido'

dataSet['ID_Pedido'].str.split('-')
```

```python
dataSet['ID_Pedido'].str.split('-').str[1].head()
```

```python
# Criando coluna no pandas

dataSet['Ano_Compra'] = dataSet['ID_Pedido'].str.split('-').str[1]
```

```python
dataSet.head()
```

```python
dataSet.Ano_Compra.value_counts()
```
