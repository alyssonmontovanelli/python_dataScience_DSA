# Criando Gr√°ficos com SEABORN

```python
# Carregando datasets do seaborn
dados = sea.load_dataset("tips") 
#Restaurante
```

```python
dados.head()
```

```python
dados.describe()
```

```python
# joinplot cria plot de 2 variaveis - graficos bi e univariado
sea.jointplot(data = dados, x = 'total_bill', y= 'tip', kind = 'reg')
```

```python
# Method implot() - plot com dados e modelos de regress√£o
sea.lmplot(data = dados, x = 'total_bill', y = 'tip', col = 'smoker')
```

## Construindo um dataframe com Pandas

```python
df = pd.DataFrame()
```

```python
# Alimentando o df com valores aleatorios
df['idade'] = random.sample(range(20,100), 30)
df['peso'] = random.sample(range(55,150), 30)
```

```python
df.shape
```

```python
df.head()
```

```python
# Implot

sea.lmplot(data = df, x = 'idade', y = 'peso', fit_reg = True) #modelo de regress√£o inclusive
```

```python
# Kdeplot

sea.kdeplot(df.idade)
```

```python
sea.kdeplot(df.peso)
```

```python
# displot

sea.distplot(df.peso)
```

```python
# Histograma
plt.hist(df.idade, alpha = .3)
sea.rugplot(df.idade) # "rugas no histograma para mais detalhamento"
```

```python
# Box Plot
sea.boxplot(df.idade, color = 'm')
```

```python
# Clustermap

sea.clustermap(df)
```
