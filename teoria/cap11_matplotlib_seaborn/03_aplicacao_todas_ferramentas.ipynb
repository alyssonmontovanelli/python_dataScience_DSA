# Usando Matplotlib, Seaborn, NumPy e Pandas

```python
import random
import numpy as np
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt
import warnings
import seaborn as sea
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
np.random.seed(42)
n = 1000
pct_fumantes = 0.2

# Variáveis
fumantes = np.random.rand(n) < pct_fumantes
idade = np.random.random_integers(15,90,n)
altura = np.random.normal(170, 10, n).round(2)
peso = np.random.normal(70, 10, n).round(2)

# DataFrame
dataSet = pd.DataFrame({'Altura': altura, 'Peso': peso, 'Fumante': fumantes})

```

```python
altura.min()
```

```python
dataSet.head()
```

```python
# Criando dados para fumante

dataSet['Fumante'] = dataSet['Fumante'].map({True: 'Fumante', False: 'Não Fumante'})
```

```python
dataSet.head()
```

```python
sea.set(style = 'ticks')

sea.lmplot(x = 'Altura',
           y = 'Peso',
           data = dataSet,
           hue = 'Fumante',
           palette = ['tab:blue', 'tab:red'],
           height = 7)

plt.xlabel('Altura (cm)')
plt.ylabel('Peso (Kg)')
plt.title('Relação Altura/Peso de Fumantes e Não Fumantes')
plt.show()
```
